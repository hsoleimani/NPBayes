# Variational Inference for Dirichlet Process Mixtures

This code implements the variational inference algorithm for DPGMMs described in the paper ([Blei and Jordan 2005](http://www.cs.columbia.edu/~blei/papers/BleiJordan2004.pdf)). We first generate data from a mixture of Gaussians and then try to learn a DPGMM with fixed inverse (diagonal) covariance matrices. So, the only unknown parameter of the mixture components is the mean.

First, generating the data: (trying to replicate the data shown in Fig. 2 of the paper.)

**Testing VI in a DP mixture of Gaussians with fixed inverse covariance matrix $\Lambda$**

We generate synthetic data from four 2-dimensional Gaussian distributions. 

$$f(x_n|\mu_i,\Lambda) = \frac{|\Lambda|^{1/2}}{2\pi}\exp(-\frac{1}{2}(x_n-\mu_i)\Lambda(x_n-\mu_i))$$

We also place Gaussian prior on the mean $\mu_i$; i.e. $p(\mu_i|\mu_0,\Lambda_0) = \mathcal{N}(\mu_0,\Lambda^{-1}_0)$.

The variational lower bound is:

$$Q = E_q[\log p(v|\alpha)] + E_q[\log p(\mu|\Lambda_0)] + \sum_{n=1}^N \bigg[E_q[\log p(z_n|v)] + E_q[\log p(x_n|z_n)]\bigg] - E_q[\log q(v|\gamma)] - E_q[\log q(\mu|\tau)] - \sum_{n=1}^{N}E_q[\log q(z_n|\phi_n)],$$

where $v_i\sim Beta(1,\alpha)$, $z_n \sim Mult(\pi)$, $\pi_i = v_i\prod_{j=1}^{i-1}(1-v_j)$, and $q(\mu_i|\tau_i) = \mathcal{N}(\mu_i;\tau_i,\Lambda_0)$.

As ususal in variational inference for DP, we need to choose a parameter for the maximum number of possible components $T$. Here, we also test the effect of $T$.

All the expectations are trivial. The only term that requires a little work is $E_q[p(z_n|v)]$.

$p(z_n|v) = \prod_{i=1}^{\infty}\big[v_i\prod_{j=1}^{i-1}(1-v_j)\big]^{z_{ni}}$. $\log p(z_n|v) = \sum_i I(z_n=i)\log(v_i) + \sum_i I(z_n=i)\sum_{j=1}^{i-1}\log(1-v_j)$. 
Suppose $z_n = i^*$, then $\sum_i I(z_n=i)\sum_{j=1}^{i-1}\log(1-v_j) = \sum_{j=1}^{i^*-1}\log(1-v_j) = \sum_{j=1}^{\infty}I(z_n>j)\log(1-v_j)$. Therefore, $\log p(z_n|v) = \sum_{i=1}^{\infty}[I(z_n=i)\log(v_i) + I(z_n>i)\log(1-v_i)]$. Thus, 
$$E_q[\log p(z_n|v)] = \sum_{i=1}^{\infty}[q(z_n=i)\log(v_i) + q(z_n>i)\log(1-v_i)]$$
where $q(z_n=i) = \phi_{n,i}$ and $q(z_n>i) = \sum_{j=i+1}^{T}\phi_{n,j}$.

After taking the expectations, the lowr bound $Q$ is:
\begin{align}
Q &= (\alpha-1)\sum_{i=1}^T \big(\Psi(\gamma_{i_2}) - \Psi(\gamma_{i_1}+\gamma_{i_2})\big) - \frac{1}{2}\sum_{i=1}^{T}\bigg[tr(\Lambda_0(\Lambda^{-1}+\tau_i\tau^T_i))-2\tau^T_i\Lambda_0\mu_0 \bigg] \\
&+ \sum_{n}\sum_i\bigg[\big(\Psi(\gamma_{i_2}) - \Psi(\gamma_{i_1}+\gamma_{i_2})\big)\sum_{j=i+1}^{T}\phi_{n,j} + \phi_{nj}\big(\Psi(\gamma_{i_1}) - \Psi(\gamma_{i_1}+\gamma_{i_2})\big)\bigg] \\
&+ \sum_{n}\sum_i \phi_{ni}\bigg[\tau^T_i\Lambda x_n -0.5tr(\Lambda(\Lambda^{-1}+\tau_i\tau^T_i))\bigg]\\
&- \frac{1}{2}\sum_i\log(|\Lambda|) - \sum_n\sum_i\phi_{ni}\log(\phi_{ni})\\
&- \sum_i\bigg[\Psi(\gamma_{i_1}+\gamma_{i_2}) - \Psi(\gamma_{i_1}) -\Psi(\gamma_{i_2}) + \sum_{k=1}^{2}(\gamma_{ik}-1)\big[\Psi(\gamma_{i_k}) - \Psi(\gamma_{i_1}+\gamma_{i_2})\big]\bigg].
\end{align}

The variational parameters are updated as follows:

$$\gamma_{i1} = 1+\sum_{n}\phi_{ni}$$
$$\gamma_{i2} = \alpha+\sum_{n}\sum_{j=i+1}^{T}\phi_{nj}$$
$$\phi_{ni}\propto \exp\bigg[ \Psi(\gamma_{i1}) + \sum_{t=1}^{i-1}\big(\Psi(\gamma_{t_2}) - \Psi(\gamma_{t_1}+\gamma_{t_2}) \big)  + \tau^T_i\Lambda x_n - \frac{1}{2}tr(\Lambda(\Lambda^{-1}+\tau_i\tau^T_i))\bigg]$$
$$\tau_i = \big(\Lambda_0 + \Lambda\sum_n\phi_{ni}\big)^{-1}\big(\Lambda_0\mu_0 + \Lambda\sum_{n}\phi_{ni}x_n\big)$$

```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.cm as cm
%matplotlib inline

np.random.seed(1415117801)

K = 4 # four components
mu = np.zeros((2,4))
sigma = 5.0*np.eye(2)
mu[0,:] = [-15., 0., 15., -15.]
mu[1,:] = [-40., 0., 20., 50.]
alpha = np.array([0.35,0.2,0.35,0.1]) # mixing proportions
N = 100 # number of samples

X = np.zeros((N,2))
for n in range(N):
    z = np.random.choice(4, p = alpha)
    X[n,:] = np.random.multivariate_normal(mu[:,z],sigma)
    

# plot data
def plot_data(data,mu,sigma,alpha):
    
    xmin = np.min(data[:,0])
    xmax = np.max(data[:,0])
    ymin = np.min(data[:,1])
    ymax = np.max(data[:,1])
    K = len(alpha)
    delta = 0.25
    x = np.arange(xmin, xmax, delta)
    y = np.arange(ymin, ymax, delta)
    X, Y = np.meshgrid(x, y)
    k = 0
    Z = alpha[k]*mlab.bivariate_normal(X, Y, sigma[0,0], sigma[1,1], mu[0,k], mu[1,k], sigma[1,1])
    for k in range(1,K):
        Z += alpha[k]*mlab.bivariate_normal(X, Y, sigma[0,0], sigma[1,1], mu[0,k], mu[1,k], sigma[1,1])
    
    plt.figure()
    plt.plot(data[:,0],data[:,1],'o')
    #plt.imshow(Z, interpolation='bilinear', origin='lower',cmap = cm.Greens)
    plt.contour(X,Y,Z)
    plt.show()
    

plot_data(X,mu,sigma,alpha)
```

```python
T = 20 # truncation level
alpha = 0.1

mu0 = np.mean(X,0)
Lambda0 = 1.0/np.diag(sigma)
Lambda = 1.0/np.diag(sigma)

Maxiter = 1000
ConvTH = 5e-4

from scipy.special import digamma

           

# variational parameters
tau = np.zeros((2,T))
gamma = np.zeros((2,T))
phi = np.zeros((N,T))
tauss = np.zeros((2,T))
phiss = np.zeros((2,T))

phi[:] = 1.0/T
phiss[0,:] = np.sum(phi,0)
phiss[1,:] = [np.sum(phiss[0,(i+1):(T+1)]) for i in range (T)] 
# init tau
from sklearn.cluster import KMeans
km = KMeans(n_clusters=T)
km.fit(X)
tau = km.cluster_centers_.T
#for t in range(T):
#    ind = np.random.choice(N,N/T)
#    tau[:,t] = np.mean(X[ind,:],0)

# init var parameters
gamma[0,:] = 1.0 + phiss[0,:]
gamma[1,:] = alpha + phiss[1,:]
A = np.zeros(T)


#initial plot
vhat = gamma[0,:]/np.sum(gamma,0)
pihat = [vhat[i]*np.prod(1.0-vhat[0:i]) for i in range(len(vhat))]
pihat /= np.sum(pihat)
print('Initial')
plt.figure()
plt.plot(pihat,'-*')
plt.show()
plot_data(X,tau,sigma,pihat)


# VI-iterations
variter = 0
prev_varlkh = 0.0;
varlkh = 0.0
N = X.shape[0]



while True:
    tauss[:] = 0.0;
    # loop over samples
    for n in range(N):
        
        #phiss[0,:] -= phi[n,:]
        #phiss[1,:] -= [np.sum(phi[n,(i+1):(T+1)]) for i in range (T)] 
        
        psigamma = digamma(gamma)
        psi_sumgamma = digamma(np.sum(gamma,0))
        A = psigamma[0,:] + [np.sum(psigamma[1,0:i]-psi_sumgamma[0:i]) for i in range(T)]
        temp = tau*Lambda.reshape(-1,1)
        A += np.dot(X[n,:]-0.5, temp) - 0.5*np.sum(temp*tau,0)
        A -= np.max(A)
        phi[n,:] = np.exp(A)
        phi[n,:] /= np.sum(phi[n,:])
        
        # update ss
        #phiss[0,:] += phi[n,:]
        #phiss[1,:] += [np.sum(phi[n,(i+1):(T+1)]) for i in range (T)] 
        phiss[0,:] = np.sum(phi,0)
        phiss[1,:] = [np.sum(phiss[0,(i+1):(T+1)]) for i in range (T)] 
        tauss[0,:] += phi[n,:]*X[n,0]
        tauss[1,:] += phi[n,:]*X[n,1]
        
        # update gamma
        gamma[0,:] = 1.0 + phiss[0,:]
        gamma[1,:] = alpha + phiss[1,:]
        
        temp = tau*Lambda.reshape(-1,1)
        temp2 = np.dot(X[n,:], temp) - 0.5*np.sum(temp*tau,0) - np.log(phi[n,:]+1e-50)
        varlkh +=  np.dot(phi[n,:], temp2)
    
    temp = tau*Lambda.reshape(-1,1)
    varlkh -= 0.5*np.sum(np.sum(temp2*tau,0) - 2*np.dot(mu0, temp))
    
    # update tau
    for d in range(2):
        tau[d,:] = (Lambda0[d]*mu0[d] + Lambda[d]*tauss[d,:])/(Lambda0[d] + Lambda[d]*phiss[0,:])
    
    conv = np.fabs(varlkh-prev_varlkh)/np.fabs(prev_varlkh)
    if (varlkh < prev_varlkh):
        print('oooops')
    if (conv < ConvTH) or (variter > Maxiter):
        break
    variter += 1
    prev_varlkh = varlkh
    
    if variter%500 == 0:
        vhat = gamma[0,:]/np.sum(gamma,0)
        pihat = [vhat[i]*np.prod(1.0-vhat[0:i]) for i in range(len(vhat))]
        pihat /= np.sum(pihat)
        print('step %d' %variter)
        plot_data(X,tau,sigma,pihat)
        

print('Final')
vhat = gamma[0,:]/np.sum(gamma,0)
pihat = [vhat[i]*np.prod(1.0-vhat[0:i]) for i in range(len(vhat))]
pihat /= np.sum(pihat)
plot_data(X,tau,sigma,pihat)
```
