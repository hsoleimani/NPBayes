# Tutorial on Dirchlet Process Mixture Models (DPMM)

We begin by reviweing Dirichlet Processes (DP) based on [Dirichlet Process (Teh 2006)](http://www.gatsby.ucl.ac.uk/~ywteh/research/npbayes/dp.pdf) and doing some experiments to better udnerstand DPs.

**Definition of DP**: 

A random distribution G is distributed according to a DP if all of its finite marginals are distributed according to a Dirichlet distribution.  

Specifically, if $G \sim DP(\alpha, H)$ where $\alpha$ is a positive number (concentration parameter) and $H$ is a distribution over $\Theta$, then for any finite measurable partition $A_1, A_2,...,A_r$ of $\Theta$, the vector $(G(A_1),...,G(A_r))$ is random with Dirichlet distribution:

$$(G(A_1),...,G(A_r)) \sim Dir(\alpha H(A_1),...,\alpha H(A_r)).$$

Also, $E[G(A)] = H(A)$ and $Var[G(A)] = H(A)(1-H(A))/(\alpha + 1)$.


**Posterior Distribution**

Suppose $G\sim DP(\alpha, H)$ and $\theta_1,...,\theta_n \sim G$, $\theta_i$ iid. What is the posterior distribution of $G$ given  $\theta_1,...,\theta_n$?

Let $A_1, ..., A_r$ be a finite measurable partition of $\Theta$ and let $n_k = \# \{i: \theta_i \in A_k\}$. Then, it is easy to see that $n_1,..., n_k$ are distributed according to a Multinomial distribution with parameter $(G(A_1),...,G(A_r))$. Using conjugacy between Multinomial and Dirichlet distributions, we conclude that:

$$(G(A_1),...,G(A_r))|\theta_1,...,\theta_r \sim Dir(\alpha H(A_1)+n_1, ..., \alpha H(A_r)+n_r).$$

Since, this is true for every finite measurable partition, then the posterior itself must be a DP:

$$G|\theta_1,...,\theta_r \sim DP(\alpha+n, \frac{\alpha}{\alpha+n}H + \frac{n}{\alpha+n}\frac{1}{n}\sum_{i=1}^{n}\delta_{\theta_i}).$$

**Predictive Distribution**

What is the predictive distribution of $\theta_{n+1}$ conditioned on $\theta_1,...,\theta_n$ and with $G$ marginalized out? For a measurable $A \subset \Theta$, we are interested in:

$$p(\theta_{n+1}\in A|\theta_1,...,\theta_n) = \int{p(\theta_{n+1}\in A,G|\theta_1,...,\theta_n)dG} = \int{p(\theta_{n+1}\in A|G)p(G|\theta_1,...,\theta_n)dG} = \int{G(A)p(G|\theta_1,...,\theta_n)dG} = E[G(A)|\theta_1,...,\theta_n].$$ 

And, $p(\theta_{n+1}\in A|\theta_1,...,\theta_n)=E[G(A)|\theta_1,...,\theta_n] = \frac{1}{\alpha+n}(\alpha H(A)+\sum_{i=1}^{n}\delta_{\theta_i}(A))$. Therefore, 

$$\theta_{n+1}|\theta_1,...,\theta_{n} \sim \frac{1}{\alpha + n}\big(\alpha H + \sum_{i=1}^{n}\delta_{\theta_i}\big).$$

Let $\theta^*_1, ..., \theta^*_m$ be the unique values among $\theta_1,...,\theta_n$ and $n_k$ the number of repeats of $\theta^*_k$, then the predictive distribution will be:

$$\theta_{n+1}|\theta_1,...,\theta_n \sim \frac{1}{\alpha + n}\big(\alpha H + \sum_{k=1}^{m}n_k\delta_{\delta^*_k}\big).$$

**Chinese Restaurant Process**

From the predictive distribution, it is easy to understand the chinese restaurant process. The ($n+1$)st customer either joins an already occupied table $k$ with probability proportional to $n_k$ (number of customers at that table) or sits at a new table with probability proportional to $\alpha$. The code below, generates samples from CRP.

```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(1415117801)
```

```python
# A sample from Chinese Restaurant Process (CRP)

N = 3000 # total number of samples
alpha = 2.95; # concentration parameter
zn = np.zeros(N) # cluster (table) that each data point is assigned to
numz = 0 # number of clusters already selected
nklist = list() # a list containing number of points assigned to each cluster

# 1st sample generate a cluster
numz = 1
nklist.append(1)
zn[0] = 0
for n in range(1,N):
    u = np.random.random()
    if u < alpha/(alpha + n): # create a new cluster
        nklist.append(1)
        zn[n] = numz
        numz += 1
    else: # choose from one of the existing clusters
        z = np.random.choice(numz, 1, p = np.array(nklist)/float(n))[0]
        zn[n] = z
        nklist[z] += 1
        
print(nklist)
counts, bins, patches = plt.hist(zn)
```

As we increase $\alpha$, the number of clusters increases. In fact, at step $i$, independently on the number of existing clusters, we may create a new cluster with probability $\frac{\alpha}{\alpha + i - 1}$. Therefore, we can take expectation of the number of clsuters $m$ conditioned on the sample size $n$: (sum of $n$ independent trials with success probability $\frac{\alpha}{\alpha + i - 1}$)

$$E[m|n] = \sum_{i=1}^{n}{\frac{\alpha}{\alpha + i -1}} \approx \alpha \log(1+\frac{n}{\alpha})$$ 

We now sample from a DP with a Gaussian base. This is a nice example because we can see that as we increase $\alpha$, the variance of DP samples approaches 0 and since $E[G]=H$, the histogram of samples will become closer to normal distribution.

```python
def chinese_restaurant_process(N, alpha):
    
    theta0 = np.zeros(N) # theta sampled for each data point 
    # when we assign a sample to a cluster, it inherits the theta of that cluster
    numz = 0 # number of clusters already selected
    nklist = list() # a list containing number of points assigned to each cluster
    thetalist = list() # list containing theta0 of each cluster
    # 1st sample generate a cluster
    numz = 1
    nklist.append(1)
    thetalist.append(np.random.normal())
    theta0[0] = thetalist[0]
    alpha = float(alpha)
    for n in range(1,N):
        u = np.random.random()
        if u < alpha/(alpha + n): # create a new cluster
            nklist.append(1)
            thetalist.append(np.random.normal())
            theta0[n] = thetalist[numz]
            numz += 1
        else: # choose from one of the existing clusters
            z = np.random.choice(numz, 1, p = np.array(nklist)/float(n))[0]
            theta0[n] = thetalist[z]
            nklist[z] += 1
            
    return(numz, theta0)


plt.figure()
plt.subplot(131)
nc, theta = chinese_restaurant_process(4000, 1)
counts, bins, patches = plt.hist(theta,normed = True)
plt.title('alpha = 1')
plt.subplot(132)
nc, theta = chinese_restaurant_process(4000, 10)
counts, bins, patches = plt.hist(theta,normed = True)
plt.title('alpha = 10')
plt.subplot(133)
nc, theta = chinese_restaurant_process(4000, 100)
counts, bins, patches = plt.hist(theta,normed = True)
plt.title('alpha = 100')
```

To demonstrate that $E[G] = H$, we sample $T = 2000$ times from a CRP with Gaussian base and each time we draw $N = 4000$ samples. Then, we look at the histogram of all $NT$ draws. The overall distribution as we expected is close to normal. 

```python
# sample T = 1000 times and take average
T = 200
N = 4000
theta = np.zeros(N*T)
numz = np.zeros(T)
for t in range(T):
    nc, theta0 = chinese_restaurant_process(N, 0.6)
    theta[t*N:(t+1)*N] = theta0
    numz[t] = nc

plt.subplot(121)
counts, bins, patches = plt.hist(theta,normed = True)
plt.title('H')
plt.subplot(122)
counts, bins, patches = plt.hist(numz,normed = True)
plt.title('# clusters')
```

**Stick-breaking construction of DP**

Here, we first sample $\nu_k \sim Beta(1,\alpha)$ and compute cluster proportions by $\beta_k = \nu_k\prod_{l=1}^{k-1}(1-\nu_l)$. Then $G = \sum_{k=1}^{\infty}\beta_k \delta_{\theta^*_k}$, where $\theta^* \sim H$.

```python
def stick_breaking_process(N, alpha):
    
    alpha = float(alpha)
    theta0 = np.zeros(N) # theta for each sampple
    beta = np.zeros(N) # weight for each sample
    # 1st sample generate a cluster
    nu = np.random.beta(1,alpha)
    beta[0] = nu
    theta0[0] = np.random.normal()
    prodagg = 1 - nu
    for n in range(1,N):
        nu = np.random.beta(1,alpha)
        beta[n] = nu*prodagg
        prodagg = prodagg*(1-nu)
        theta0[n] = np.random.normal()
           
    return(beta, theta0)

beta, theta0 = stick_breaking_process(4000,100)
#print(sum(beta))
#plt.subplot(121)
plt.bar(theta0, beta, width=0.01)
#plt.subplot(122)
#plt.plot(np.log(beta+1e-20))
```

**Dirichlet Process Mixture Model**

This part is based on two papers: 1. Markov chain sampling methods for dirichlet process mixture models, [Neal 2000](http://www.stat.columbia.edu/npbayes/papers/neal_sampling.pdf). 2. Dirichlet process gaussian mixture models: Choice of the base distribution, [Rasmussen 2010](http://mlg.eng.cam.ac.uk/pub/pdf/GoeRas10.pdf).

We observe $x_1,x_2,...,x_n$ which are independent and exchangable. We assume that $x_i$ is generated from a mixture of distributions of form $F(\theta)$ with the mixing proportions over $\theta$ being a Dirichlet process $G$; i.e.

$x_i|\theta_i \sim F(\theta_i)$, $\theta_i|G \sim G$, and $G \sim DP(G_0,\alpha)$.

To derive posterior distributions of this model, we first assume that we have a mixture model with finite number of $K$ compoenents and then take the limit as $K$ goes to infinity. According to (Neal 2000) this approach is usually useful for deriving algorithms for sampling from the posterior of DP mixture models. Also, (Neal 2000) claims that an algorithm for DP mixture model usually have a counterpart for finite mixture models. Accordingly, we assume that:

$x_i|z_i,\phi \sim F(\phi_{z_i})$, $z_i|\pi \sim Discrete(\pi_1,...,\pi_K)$, $\phi_z \sim G_0$, and $\pi \sim Dirichlet(\alpha/K,...,\alpha/K)$.

By integrating out $\pi$, it is easy to see that $p(z_i = j|others) = \frac{n_{-i,j}+\alpha/K}{n-1+\alpha}$, where $n_{-i,j} = \sum_{i'\neq i}\delta(z_{i'}=j)$. Now, by letting $K\rightarrow \infty$, we have $p(z_i = j|others) \rightarrow \frac{n_{-i,j}}{n-1+\alpha}$. Therefore,

$$p(z_i=j|others) = \frac{n_{-i,j}}{n-1+\alpha}$$
$$p(z_i\neq z_{i'}\forall i'\neq i|others) = \frac{\alpha}{n-1+\alpha}$$

Therefore, for each sample $x_i$, the conditional distribution of $\theta_i$ conditioned on all other paramters is obtained by combining the likelihood of $x_i$ and $\theta_i$ ($F(x_i,\theta_i)$) and the prior of $\theta_i$ conditioned on $\theta_{-i}$. The latter part is:

$$\theta_i|\theta_{-i} \sim \frac{1}{n-1+\alpha}\sum_{j\neq i}\delta(\theta_j) + \frac{\alpha}{n-1+\alpha}G_0$$

By multiplying this expression by $F(x_i,\theta_i)$, we obtain:

$$\theta_i|\theta_{-i},x_i \propto \sum_{j\neq i}F(x_i,\theta_i)\delta(\theta_i) + \alpha\int F(x_i,\theta)dG_0(\theta)$$

The integral in the second term is analytically computable if $G_0$ is a conjugate prior for the likelihood.

In other words, the posterior of latent variable $z_i$ conditioned on all other parameters is:

$$p(z_i=j|z_{-i},x_i,\phi) \propto \frac{n_{-i,j}}{n-1+\alpha}F(x_i, \phi_j),~if z_{i'}=j~ \text{for some}~ i'\neq i$$
$$p(z_i=j|z_{-i},x_i,\phi) \propto \frac{\alpha}{n-1+\alpha}\int F(x_i, \phi)dG_0(\phi),~if z_{i'}\neq j~ \forall i'\neq i$$


**Dirichlet Process Gaussian Mixture Model**

See the "GMM\_GibbsSampling" notebook for a simple 2-component GMM. Here, we extend that model to a DPGMM. 

*Model setup:*

$p(x_d|\Theta) = \sum_{j=1}^{K}\pi_j\prod_{n=1}^{N}\mathscr{N}(x_{dn}|\mu_{jn},\sigma^2_{jn})$, $x_d \in R^{N}$, $\forall d=1,2,...,D.$

$p(x_{d}|z_d,\Theta) = \prod_{j=1}^{K}\bigg[\prod_{n=1}^{N}\mathscr{N}(x_{dn}|\mu_{jn},\sigma^2_{jn})\bigg]^{z_{dj}}$, $p(z_d|\pi) = \prod_{j=1}^{K}\pi^{z_{dj}}_j$, $K \rightarrow \infty$.

Prior on $\pi$: $p(\pi_1,...,\pi_K|\alpha) = \frac{\Gamma(\alpha)}{\Gamma(\alpha/K)^K}\prod_{j=1}^{K}\pi^{\alpha/K-1}_j$, $K \rightarrow \infty$.

Prior on $\mu_{jn}$ and $\sigma^2_{jn}$: $(\mu_j,S_j)\sim NIG(\xi_{jn},\rho_{jn},a_{jn},b_{jn})$ where:
$$(\mu_{jn}|\sigma^2_{jn}, \xi_{jn}, \rho_{jn}) \sim \mathscr{N}(\xi_{jn}, \frac{\sigma^2_{jn}}{\rho_{jn}}) = \frac{\sqrt{\rho_{jn}}}{\sqrt{2\pi\sigma^2_{jn}}}\exp(-\frac{\rho_{jn}}{2\sigma^2_{jn}}(\mu_{jn}-\xi_{jn})^2)$$
$$(\sigma^2_{jn}|a_{jn},b_{jn}) \sim Gamma^{-1}(a_{jn},b_{jn}) = \frac{b^{a_{jn}}_{jn}}{\Gamma(a_{jn})}\sigma^{-2(a_{jn}+1)}exp(b_{jn}/\sigma^2)$$

Priors on the hyper-parameters:

$\xi_{jn} \sim \mathscr{N}(\mu_0,\sigma^2_0)$, $\rho_{jn} \sim \Gamma(1,1)$,
$(a_{jn}-1) \sim \Gamma(1,1)$, and $b_{jn} \sim \Gamma(\sigma^2_0,1)$ $\forall j,n$
where $\mu_0$ and $\sigma^2_0$ are respectively empirical mean and variance of the data.

*Gibbs Sampling Algorithm:*

- update mixture parameters $(\mu_{jn}, \sigma^2_{jn})$ conditioned on latent variables $z$ and hyper-parameters
- update hyper-parameters $\xi_{jn}$, $\rho_{jn}$, $b_{jn}$, and $a_{jn}$ conditioned on other parameters
- update latent variables $z$: For $d = 1,...,D$:
    - if $n_{-d,j} > 0$: $p(z_d=j|z_{-d},x_d,\Theta) \propto \frac{n_{-d,j}}{n-1+\alpha}\prod_{n=1}^{N}\mathscr{N}(x_{dn};\mu_{jn},\sigma^2_{jn})$
    - else: $p(z_d=j|z_{-d},x_d,\Theta) \propto \frac{\alpha}{n-1+\alpha}\prod_{n=1}^{N}\Big[\int \mathscr{N}(x_{dn};\mu_{\cdot n},\sigma^2_{\cdot n})NIG(\mu_{\cdot n},\sigma^2_{\cdot n}|\xi_{jn}, \rho_{jn},b_{jn}, a_{jn})d\mu_{\cdot n}d\sigma^2_{\cdot n}\Big]$
    

Note that $\int \mathscr{N}(x_{dn};\mu_{\cdot n},\sigma^2_{\cdot n})NIG(\mu_{\cdot n},\sigma^2_{\cdot n}|\xi_{jn}, \rho_{jn},b_{jn}, a_{jn})d\mu_{\cdot n}d\sigma^2_{\cdot n} = t_{2a}\bigg(x_{dn}|\xi_{jn}, \lambda = \frac{a_{jn}\rho_{jn}}{b_{jn}(1+\rho_{jn})}\bigg)$

```python
data = np.loadtxt('faithful.txt')[:,1:]
plt.plot(data[:,0],data[:,1],'o')
```

```python
# First initialize the model with only 1 component
# As new components are created, we extend parameter vectors

from scipy.stats import invgamma
from scipy.stats import norm
from scipy.special import gammaln
from scipy.stats import t as tdistn

np.random.seed(1419)

D = data.shape[0] # sample size
N = data.shape[1] # dimension
z = np.zeros(D,dtype=int) # hidden variable

# initialize hyper-paramters
mu0 = np.mean(data,0)
sigma02 = np.diag(np.cov(data.T))
xi = np.zeros((N,1)) # hyper-parameter on mu
rho = np.zeros((N,1)) # hyper-parameter on mu
a = np.zeros((N,1)) # hyper-parameter on sigma2
b = np.zeros((N,1)) # hyper-parameter on sigma2
for n in range(N):
    xi[n] = np.random.normal(mu0[n], np.sqrt(sigma02[n]), 1)
    rho[n] = np.random.gamma(1,1.0/1, 1)
    a[n] = 1.0 + np.random.gamma(1,1, 1)
    b[n] = np.random.gamma(sigma02[n], 1, 1)

alpha = .1 # Dirichlet hyper-parameter

# first run to initialize cluster
numz = 0 # number of clusters already selected
nklist = list() # a list containing number of points assigned to each cluster
# 1st sample generate a cluster
K = 1
nj = np.array([1])
z[0] = 0
alpha = float(alpha)
mu = np.zeros((N,K)) # mean of component k dimension n
sigma2 = np.zeros((N,K)) # variance of component k dimension n
for n in range(N):
    sigma2[n,0] = invgamma.rvs(a[n],scale=b[n])
    mu[n,0] = np.random.normal(xi[n],np.sqrt(sigma2[n,0]/rho[n]))

for d in range(1,D):
    activeind = np.where(nj > 0)[0]
    activeind = np.hstack((activeind,np.array([K])))
    logprobvec = np.zeros(K+1)
    for j in activeind:
        if j!=K:
            logprobvec[j] = np.log(nj[j]) + np.sum(norm.logpdf(data[d,:], loc = mu[:,j], scale = np.sqrt(sigma2[:,j])))
        else:
            lamda = a*rho/(b*(1+rho))
            logprobvec[j] = np.log(alpha) + np.sum(tdistn.logpdf(data[d,:].reshape(-1,1), 2*a,loc=xi, scale=np.sqrt(1/lamda)))
    maxlog = max(logprobvec[activeind])
    probvec = np.zeros(K+1)
    probvec[activeind] = np.exp(logprobvec[activeind] - maxlog)
    z[d] = np.random.choice(activeind,1,p = probvec[activeind]/np.sum(probvec))[0]
    if z[d] < K:
        nj[z[d]] += 1
    else:
        nj = np.hstack((nj,np.array([1])))
        K += 1
        # sample new mu and sigma2
        sigma2 = np.hstack((sigma2,np.zeros((N,1))))
        mu = np.hstack((mu,np.zeros((N,1))))
        j = K-1
        for n in range(N):
            sigma2[n,j] = invgamma.rvs(a[n],scale=b[n])
            mu[n,j] = np.random.normal(xi[n],np.sqrt(sigma2[n,j]/rho[n]))


import matplotlib.mlab as mlab
def contour_plot(data,z,mu,sigma,nj):
    z.dtype = int
    delta = 0.1
    colors = ['r','b','g','k','y','r','b','g','k','y']
    xlim = [min(data[:,0]),max(data[:,0])]
    ylim = [min(data[:,1]),max(data[:,1])]
    x = np.arange(xlim[0]*0.5, xlim[1]*1.5, delta)
    y = np.arange(ylim[0]*0.5, ylim[1]*1.5, delta)
    X, Y = np.meshgrid(x, y)
    #matplotlib.mlab.bivariate_normal(X, Y, sigmax=1.0, sigmay=1.0, mux=0.0, muy=0.0, sigmaxy=0.0)
    # difference of Gaussians
    #Z = 10.0 * (Z2 - Z1)
    plt.figure()
    for i,j in enumerate(np.where(nj > 0)[0]):
        Z1 = mlab.bivariate_normal(X, Y, sigma[0,j], sigma[1,j], mu[0,j], mu[1,j], 0.0)
        CS = plt.contour(X, Y, Z1,colors=colors[i])
        plt.plot(data[z==j,0],data[z==j,1],'o',color=colors[i])
    plt.show()
    
print('# init clusters %d' %K)
contour_plot(data,z,mu,sigma2, nj)
```

```python
T = 500 # MC sample size
MCmu = mu.copy()
MCsigma2 = sigma2.copy()
MCrho = rho.copy()
MCxi = xi.copy()
MCz = z.copy()
MCz.dtype = float
MCb = b.copy()
MCa = a.copy()
MCK = np.zeros(T) # keeps number of active components in each MC step


MCK[0] = K
for t in range(1,T):
    if (t+1)%100 == 0:
        print('MC step %d' %((t+1)))
        contour_plot(data,MCz,MCmu,MCsigma2, nj)
    # update mixture parameters:
    nj = np.zeros(K)     
    for j in range(K):
        # compute ss
        nj[j] = sum(z==j)
        if nj[j] == 0:
            continue
        for n in range(N):
            xbar = np.mean(data[z==j,n],0)
            varbar = np.var(data[z==j,n].T)
            rhoss = rho[n] + nj[j]
            ass = a[n] + 0.5*nj[j]
            xiss = (rho[n]*xi[n]+nj[j]*xbar)/rhoss
            bss = b[n] + 0.5*nj[j]*varbar + 0.5*nj[j]*rho[n]*((xbar-xi[n])**2)/rhoss
            # sample sigma2
            sigma2[n,j] = invgamma.rvs(ass,scale=bss)
            # sample mu
            mu[n,j] = np.random.normal(xiss,np.sqrt(sigma2[n,j]/rhoss))
            # update MC samples
            MCsigma2[n,j] = (t*MCsigma2[n,j] + sigma2[n,j])/(t+1.0)
            MCmu[n,j] = (t*MCmu[n,j] + mu[n,j])/(t+1.0)
        
    # update hyper-parameters
    for n in range(N):
        # update xi
        sigmass = 1.0/(rho[n]*np.sum(1.0/sigma2[n,:])+1/sigma02[n])
        muss = (rho[n]*np.sum(mu[n,:]/sigma2[n,:])+mu0[n]/sigma02[n])/sigmass
        xi[n] = np.random.normal(muss, scale = np.sqrt(sigmass))
        MCxi[n] = (t*MCxi[n] + xi[n])/(t+1.0)
        #update rho
        temp = np.sum(((mu[n,:]-xi[n])**2)/sigma2[n,:])
        bss = (1.0+0.5*temp)
        rho[n] = np.random.gamma(K/2.0+1.0,1.0/bss)
        MCrho[n] = (t*MCrho[n] + rho[n])/(t+1.0)
        # update b
        bss = 1.0 + np.sum(1.0/sigma2[n,:])
        ass = sigma02[n] + K*a[n]
        b[n] = np.random.gamma(ass, 1.0/bss)
        MCb[n] = (t*MCb[n] + b[n])/(t+1.0)
        #update a
        xstar = np.random.normal(loc = a[n], scale = 0.1)
        if (xstar < 1):
            MCa[n] = (t*MCa[n] + a[n])/(t+1.0)
            continue
        logqstar = K*xstar*np.log(b[n])-K*gammaln(xstar)-(xstar+1)*np.sum(np.log(sigma2[n,:]))-(xstar-1)
        logq = K*a[n]*np.log(b[n])-K*gammaln(a[n])-(a[n]+1)*np.sum(np.log(sigma2[n,:]))-(a[n]-1)
        logMHratio = logqstar - logq
        logalpha = min(0, logMHratio)
        if (np.log(np.random.uniform()) < logalpha):
            a[n] = xstar
        MCa[n] = (t*MCa[n] + a[n])/(t+1.0)
    
    # update z
    for d in range(D):
        #tempnj = nj.copy()
        nj[z[d]] -= 1
        activeind = np.where(nj > 0)[0]
        activeind = np.hstack((activeind,np.array([K])))
        logprobvec = np.zeros(K+1)
        for j in activeind:
            if j!=K:
                logprobvec[j] = np.log(nj[j]) + np.sum(norm.logpdf(data[d,:], loc = mu[:,j], scale = np.sqrt(sigma2[:,j])))
            else:
                lamda = a*rho/(b*(1+rho))
                logprobvec[j] = np.log(alpha) + np.sum(tdistn.logpdf(data[d,:].reshape(-1,1), 2*a,loc=xi, scale=np.sqrt(1/lamda)))
        maxlog = max(logprobvec[activeind])
        probvec = np.zeros(K+1)
        probvec[activeind] = np.exp(logprobvec[activeind] - maxlog)
        z[d] = np.random.choice(activeind,1,p = probvec[activeind]/np.sum(probvec))[0]
        if z[d] < K:
            nj[z[d]] += 1
        else:
            nj = np.hstack((nj,np.array([1])))
            K += 1
            # sample new mu and sigma2
            sigma2 = np.hstack((sigma2,np.zeros((N,1))))
            mu = np.hstack((mu,np.zeros((N,1))))
            MCsigma2 = np.hstack((MCsigma2,np.zeros((N,1))))
            MCmu = np.hstack((MCmu,np.zeros((N,1))))
            j = K-1
            for n in range(N):
                sigma2[n,j] = invgamma.rvs(a[n],scale=b[n])
                mu[n,j] = np.random.normal(xi[n],np.sqrt(sigma2[n,j]/rho[n]))
        MCz[d] = (t*MCz[d] + float(z[d]))/(t+1.0)
        MCK[t] = len(np.where(nj>0)[0])
```

This works OK but is sensitive to $\alpha$ and initial value of cluster parameters. If initialized poorly, the MC chain may not mix well. Next, we'll try algorithm 3 of Neal 2000 which is collapsed Gibbs sampling; i.e. cluster parameters $\mu, \sigma^2$ are integrated out.


*Collapsed Gibbs Sampling Algorithm:*

- update hyper-parameters $\xi_{n}$, $\rho_{n}$, $b_{n}$, and $a_{n}$
- update latent variables $z$: For $d = 1,...,D$:
    - if $n_{-d,j} > 0$: $p(z_d=j|z_{-d},x_d,\Theta) \propto \frac{n_{-d,j}}{n-1+\alpha}\prod_{n=1}^{N}\int\big[\mathscr{N}(x_{dn};\mu_{jn},\sigma^2_{jn})NIW(\mu_{jn},\sigma^2_{jn}|x_{-d,n},\xi_n,\rho_n,b_n,a_n)d\mu_{jn}d\sigma^2_{jn}\big] = \frac{n_{-d,j}}{n-1+\alpha}\prod_{n=1}^{N}t_{2a}\bigg(x_{dn}|\hat{\xi}_{jn}, \lambda = \frac{\hat{a}_{jn}\hat{\rho}_{jn}}{\hat{b}_{jn}(1+\hat{\rho}_{jn})}\bigg)$
    - else: $p(z_d=j|z_{-d},x_d,\Theta) \propto \frac{\alpha}{n-1+\alpha}\prod_{n=1}^{N}\Big[\int \mathscr{N}(x_{dn};\mu_{\cdot n},\sigma^2_{\cdot n})NIG(\mu_{\cdot n},\sigma^2_{\cdot n}|\xi_{n}, \rho_{n},b_{n}, a_{n})d\mu_{\cdot n}d\sigma^2_{\cdot n}\Big]\frac{\alpha}{n-1+\alpha}\prod_{n=1}^{N}t_{2a}\bigg(x_{dn}|\xi_{jn}, \lambda = \frac{a_{jn}\rho_{jn}}{b_{jn}(1+\rho_{jn})}\bigg)$
    
where $\hat{\xi}_{jn} = \frac{n_j\bar{X}_{jn}+\rho_{jn}\xi_{jn}}{n_j+\rho_{jn}}$, $\hat{\rho}_{jn} = \rho_{jn} + n_j$, $\hat{a}_{jn} = a_{jn} + 0.5n_j$, and $\hat{b}_{jn} = b_{jn} + \frac{1}{2}\sum_{d:z_d=j}(x_{dn}-\bar{X}_{jn})^2 + \frac{1}{2}\frac{n_j\rho_{jn}}{n_j+\rho_{jn}}(\bar{X}_{jn}-\xi_{jn})^2$. (all excluding the current data point)

Gibbs updates for hyper-parameters:

$\mu$ and $\sigma^2$ are integrated out for each cluster. The likelihood conditioned on hyper-parameters is :

$$p(X|a,b,\rho,\xi) = \prod_{n=1}^N\prod_{j=1}^{K}\Big[\frac{b^{a_n}_n}{(2\pi)^{n_j/2}\Gamma(a_n)}\frac{\sqrt{rho_n}}{\sqrt{\rho[n]+n_j}}\frac{\Gamma(a_n+n_j/2)}{B^{a_n+n_j/2}}\Big]$$
where $B_{nj} = b_{n} + \frac{1}{2}\sum_{d:z_d=j}(x_{dn}-\bar{X}_{jn})^2 + \frac{1}{2}\frac{n_j\rho_{n}}{n_j+\rho_{n}}(\bar{X}_{jn}-\xi_{n})^2$

So, we use M-H for updating the hyper-parameters:

- $\pi(a_n)\propto e^{-a_n}\prod_{j}\frac{b^{a_n}_n}{\Gamma(a_n)}\frac{\Gamma(a_n+n_j/2)}{B^{a_n+n_j/2}}$
- $\pi(b_n)\propto e^{-b_n}b^{\sigma^2_0-1}_n\prod_{j}\frac{b^{a_n}_n}{\Gamma(a_n)}\frac{\Gamma(a_n+n_j/2)}{B^{a_n+n_j/2}}$
- $\pi(\rho_n)\propto e^{-\rho_n}\prod_{j}\frac{\sqrt{rho_n}}{\sqrt{\rho[n]+n_j}}\frac{1}{B^{a_n+n_j/2}}$
- $\pi(\xi_n)\propto \exp(-\frac{(\xi_n-\mu_0)^2}{2\sigma^2_0})\prod_{j}\frac{1}{B^{a_n+n_j/2}}$

```python
# First initialize the model with only 1 component
# As new components are created, we extend parameter vectors

from scipy.stats import invgamma
from scipy.stats import norm
from scipy.special import gammaln
from scipy.stats import t as tdistn

np.random.seed(1419)

D = data.shape[0] # sample size
N = data.shape[1] # dimension
z = np.zeros(D,dtype=int) # hidden variable

# initialize hyper-paramters
mu0 = np.mean(data,0)
sigma02 = np.diag(np.cov(data.T))
xi = np.zeros((N,1)) # hyper-parameter on mu
rho = np.zeros((N,1)) # hyper-parameter on mu
a = np.zeros((N,1)) # hyper-parameter on sigma2
b = np.zeros((N,1)) # hyper-parameter on sigma2
for n in range(N):
    xi[n] = np.random.normal(mu0[n], np.sqrt(sigma02[n]), 1)
    rho[n] = np.random.gamma(1,1.0/1, 1)
    a[n] = 1.0 + np.random.gamma(1,1, 1)
    b[n] = np.random.gamma(sigma02[n], 1, 1)

alpha = 1 # Dirichlet hyper-parameter

# first run to initialize cluster
numz = 0 # number of clusters already selected
nklist = list() # a list containing number of points assigned to each cluster
# 1st sample generate a cluster
K = 1
nj = np.array([1])
z[0] = 0
alpha = float(alpha)

for d in range(1,D):
    activeind = np.where(nj > 0)[0]
    activeind = np.hstack((activeind,np.array([K])))
    logprobvec = np.zeros(K+1)
    for j in activeind:
        if j!=K:
            xbar = np.mean(data[z[0:d]==j,:],0).reshape(N,1)
            if nj[j] == 1:
                varbar = np.zeros((N,1))
            else:
                varbar = np.diag(np.cov(data[z[0:d]==j,:].T)).reshape(N,1)
            rhoss = rho + nj[j]
            ass = a + 0.5*nj[j]
            xiss = (rho*xi+nj[j]*xbar)/rhoss
            bss = b + 0.5*nj[j]*varbar + 0.5*nj[j]*rho*((xbar-xi)**2)/rhoss
            lamda = ass*rhoss/(bss*(1+rhoss))
            logprobvec[j] = np.log(nj[j])+np.sum(tdistn.logpdf(data[d,:].reshape(-1,1), 2*ass,loc=xiss, scale=np.sqrt(1/lamda)))
        else:
            lamda = a*rho/(b*(1+rho))
            logprobvec[j] = np.log(alpha) + np.sum(tdistn.logpdf(data[d,:].reshape(-1,1), 2*a,loc=xi, scale=np.sqrt(1/lamda)))
    maxlog = max(logprobvec[activeind])
    probvec = np.zeros(K+1)
    probvec[activeind] = np.exp(logprobvec[activeind] - maxlog)
    z[d] = np.random.choice(activeind,1,p = probvec[activeind]/np.sum(probvec))[0]
    if z[d] < K:
        nj[z[d]] += 1
    else:
        nj = np.hstack((nj,np.array([1])))
        K += 1

# double check data[z[0:d]==j,:]
# check posterior mu and sigma2 and plot them
        
        
import matplotlib.mlab as mlab
def contour_plot(data,z0,mu,sigma,nj):
    delta = 0.1
    colors = ['r','b','g','k','y','r','b','g','k','y']
    xlim = [min(data[:,0]),max(data[:,0])]
    ylim = [min(data[:,1]),max(data[:,1])]
    x = np.arange(xlim[0]*0.5, xlim[1]*1.5, delta)
    y = np.arange(ylim[0]*0.5, ylim[1]*1.5, delta)
    X, Y = np.meshgrid(x, y)
    #matplotlib.mlab.bivariate_normal(X, Y, sigmax=1.0, sigmay=1.0, mux=0.0, muy=0.0, sigmaxy=0.0)
    # difference of Gaussians
    #Z = 10.0 * (Z2 - Z1)
    z = np.array([int(round(x)) for x in z0])
    plt.figure()
    for i,j in enumerate(np.where(nj > 0)[0]):
        Z1 = mlab.bivariate_normal(X, Y, sigma[0,j], sigma[1,j], mu[0,j], mu[1,j], 0.0)
        CS = plt.contour(X, Y, Z1,colors=colors[i])
        plt.plot(data[z==j,0],data[z==j,1],'o',color=colors[i])
    plt.show()
    
print('# init clusters %d' %K)
mu = np.zeros((N,K)) # mean of component k dimension n
sigma2 = np.zeros((N,K)) # variance of component k dimension n
for j in range(K):
    for n in range(N):
        xbar = np.mean(data[z==j,n],0)
        varbar = np.var(data[z==j,n].T)
        rhoss = rho[n] + nj[j]
        ass = a[n] + 0.5*nj[j]
        mu[n,j] = (rho[n]*xi[n]+nj[j]*xbar)/rhoss
        bss = b[n] + 0.5*nj[j]*varbar + 0.5*nj[j]*rho[n]*((xbar-xi[n])**2)/rhoss
        sigma2[n,j] = bss/(ass-1.0)
contour_plot(data,z,mu,sigma2, nj)
```

```python
T = 300 # MC sample size
MCmu = mu.copy()
MCsigma2 = sigma2.copy()
MCrho = rho.copy()
MCxi = xi.copy()
MCz = np.array([float(x) for x in z])
MCb = b.copy()
MCa = a.copy()
MCK = np.zeros(T) # keeps number of active components in each MC step


MCK[0] = K
for t in range(1,T):
    if (t+1)%100 == 0:
        print('MC step %d' %((t+1)))
        contour_plot(data,MCz,MCmu,MCsigma2, nj)
    # update (posterior mean) of mixture parameters:
    nj = np.zeros(K)     
    for j in range(K):
        # compute ss
        nj[j] = sum(z==j)
        if nj[j] == 0:
            continue
        for n in range(N):
            xbar = np.mean(data[z==j,n],0)
            varbar = np.var(data[z==j,n].T)
            rhoss = rho[n] + nj[j]
            ass = a[n] + 0.5*nj[j]
            xiss = (rho[n]*xi[n]+nj[j]*xbar)/rhoss
            bss = b[n] + 0.5*nj[j]*varbar + 0.5*nj[j]*rho[n]*((xbar-xi[n])**2)/rhoss
            # posterior mean of sigma2
            sigma2[n,j] = bss/(ass-1.0)
            # posterior mean of mu
            mu[n,j] = (rho[n]*xi[n]+nj[j]*xbar)/rhoss
            # update MC samples
            MCsigma2[n,j] = (t*MCsigma2[n,j] + sigma2[n,j])/(t+1.0)
            MCmu[n,j] = (t*MCmu[n,j] + mu[n,j])/(t+1.0)
        
    # update hyper-parameters
    for n in range(N):
        # update xi
        xistar = np.random.normal(loc = xi[n], scale = 0.1)
        logqstar = -((xistar - mu0[0])**2)/(2.0*sigma02[n])
        logq = -((xi[n] - mu0[0])**2)/(2.0*sigma02[n])
        for j in np.where(nj > 0)[0]:
            xbar = np.mean(data[z==j,n],0)
            varbar = np.var(data[z==j,n].T) 
            rhoss = rho[n] + nj[j]
            Bss = b[n] + 0.5*nj[j]*varbar 
            B = Bss + 0.5*nj[j]*rho[n]*((xbar-xi[n])**2)/rhoss
            Bstar = Bss + 0.5*nj[j]*rho[n]*((xbar-xistar)**2)/rhoss
            logq -= (a[n]+0.5*nj[j])*np.log(B)
            logqstar -= (a[n]+0.5*nj[j])*np.log(Bstar)
        logMHratio = logqstar - logq
        logalpha = min(0, logMHratio)
        if (np.log(np.random.uniform()) < logalpha):
            xi[n] = xistar
        MCxi[n] = (t*MCxi[n] + xi[n])/(t+1.0)
        
        # update rho
        rhostar = np.random.normal(loc = rho[n], scale = 0.1)
        if (rhostar < 0):
            MCrho[n] = (t*MCrho[n] + rho[n])/(t+1.0)
        else:
            logqstar = -rhostar
            logq = -rho[n]
            for j in np.where(nj > 0)[0]:
                xbar = np.mean(data[z==j,n],0)
                varbar = np.var(data[z==j,n].T) 
                Bss = b[n] + 0.5*nj[j]*varbar 
                temp = ((xbar-xi[n])**2)
                B = Bss + 0.5*nj[j]*rho[n]*temp/(rho[n] + nj[j])
                Bstar = Bss + 0.5*nj[j]*rhoss*temp/(rhoss + nj[j])
                logq -= (a[n]+0.5*nj[j])*np.log(B) - 0.5*np.log(rho[n]/(rho[n]+nj[j]))
                logqstar -= (a[n]+0.5*nj[j])*np.log(Bstar) - 0.5*np.log(rhoss/(rhoss+nj[j]))
            logMHratio = logqstar - logq
            logalpha = min(0, logMHratio)
            if (np.log(np.random.uniform()) < logalpha):
                rho[n] = rhostar
            MCrho[n] = (t*MCrho[n] + rho[n])/(t+1.0)
        
        # update a and b
        astar = np.random.normal(loc = a[n], scale = 0.1)
        bstar = np.random.normal(loc = b[n], scale = 0.1)
        if (astar < 1) or (bstar < 0):
            MCa[n] = (t*MCa[n] + a[n])/(t+1.0)
            MCb[n] = (t*MCb[n] + b[n])/(t+1.0)
        else:
            logqstar = - astar - bstar + (sigma02[n]-1)*np.log(bstar)
            logqstar = - a[n] - b[n] + (sigma02[n]-1)*np.log(b[n])
            for j in np.where(nj > 0)[0]:
                xbar = np.mean(data[z==j,n],0)
                varbar = np.var(data[z==j,n].T) 
                temp = 0.5*nj[j]*varbar + 0.5*nj[j]*rho[n]*((xbar-xi[n])**2)/(rho[n] + nj[j])
                logq += a[n]*np.log(b[n]) + gammaln(a[n]+0.5*nj[j]) - gammaln(a[n]) + (a[n]+0.5*nj[j])*np.log(b[n]+temp)
                logqstar += astar*np.log(bstar) + gammaln(astar+0.5*nj[j]) - gammaln(astar) + (astar+0.5*nj[j])*np.log(bstar+temp)

            logMHratio = logqstar - logq
            logalpha = min(0, logMHratio)
            if (np.log(np.random.uniform()) < logalpha):
                a[n] = astar
                b[n] = bstar
            MCa[n] = (t*MCa[n] + a[n])/(t+1.0)
            MCb[n] = (t*MCb[n] + b[n])/(t+1.0)
        
    # update z
    for d in range(D):
        #tempnj = nj.copy()
        nj[z[d]] -= 1
        activeind = np.where(nj > 0)[0]
        activeind = np.hstack((activeind,np.array([K])))
        logprobvec = np.zeros(K+1)
        for j in activeind:
            if j!=K:
                xbar = np.mean(data[z==j,:],0).reshape(N,1)
                if nj[j] == 1:
                    varbar = np.zeros((N,1))
                else:
                    varbar = np.diag(np.cov(data[z==j,:].T)).reshape(N,1)
                rhoss = rho + nj[j]
                ass = a + 0.5*nj[j]
                xiss = (rho*xi+nj[j]*xbar)/rhoss
                bss = b + 0.5*nj[j]*varbar + 0.5*nj[j]*rho*((xbar-xi)**2)/rhoss
                lamda = ass*rhoss/(bss*(1+rhoss))
                logprobvec[j] = np.log(nj[j])+np.sum(tdistn.logpdf(data[d,:].reshape(-1,1), 2*ass,loc=xiss, scale=np.sqrt(1/lamda)))
            else:
                lamda = a*rho/(b*(1+rho))
                logprobvec[j] = np.log(alpha) + np.sum(tdistn.logpdf(data[d,:].reshape(-1,1), 2*a,loc=xi, scale=np.sqrt(1/lamda)))
        maxlog = max(logprobvec[activeind])
        probvec = np.zeros(K+1)
        probvec[activeind] = np.exp(logprobvec[activeind] - maxlog)
        z[d] = np.random.choice(activeind,1,p = probvec[activeind]/np.sum(probvec))[0]
        if z[d] < K:
            nj[z[d]] += 1
        else:
            nj = np.hstack((nj,np.array([1])))
            K += 1
            # sample new mu and sigma2
            sigma2 = np.hstack((sigma2,np.zeros((N,1))))
            mu = np.hstack((mu,np.zeros((N,1))))
            MCsigma2 = np.hstack((MCsigma2,np.zeros((N,1))))
            MCmu = np.hstack((MCmu,np.zeros((N,1))))
            j = K-1
            for n in range(N):
                sigma2[n,j] = invgamma.rvs(a[n],scale=b[n])
                mu[n,j] = np.random.normal(xi[n],np.sqrt(sigma2[n,j]/rho[n]))
        MCz[d] = (float(t)*MCz[d] + float(z[d]))/(t+1.0)
        MCK[t] = len(np.where(nj>0)[0])
```
